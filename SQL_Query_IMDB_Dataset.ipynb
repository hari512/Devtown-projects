**<h1>SQL Query IMdb Dataset</h1>**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import sqlite3
```

```python
path="/content/drive/MyDrive/Data Sets/movies (1).sqlite"
con=sqlite3.connect(path)
cur=con.cursor()
```

```python
# Get all the data from table movies
cur.execute("select * from movies")
movie=cur.fetchall()
```

```python
movie=pd.DataFrame(movie,columns=['id','original_title','budget','popularity','release_date','revenue','title','vote_average','vote_count','overview','tagline','uid','director_id'])
```

```python
#Get the date from table directors
cur.execute('select * from directors')
directors=pd.DataFrame(cur.fetchall(),columns=['name', 'id', 'gender', 'uid', 'department'])
directors
```

```python
directors.info()
```

```python
# check how many movies are present in imdb data
cur.execute('select count(id) from movies')
cur.fetchall()
```

```python
# Find the data of theese three directors : James Cameron, Luc besson, John woo
cur.execute('select * from directors where name in ("James Cameron","Luc Besson","John Woo")')  # we can do using or operator
dir=pd.DataFrame(cur.fetchall(),columns=directors.columns)                                    # select * from directors where name='James Cameron' or name='Luc Besson' or name='John Woo'
dir
```

```python
# Find the all the data of directors with name starting with 'steven'
cur.execute('select name from directors where name like "steven%"')
cur.fetchall()
```

```python
# count the female directors ,gender=1
cur.execute('select count(id) from directors where gender=="1"')
cur.fetchall()
```

```python
#find the name of 10th female director.
cur.execute('select name from directors where gender=="1" limit 10')
female=cur.fetchall()
female[9]
```

```python
# What are the top 3 most popular movies.
cur.execute('select original_title from movies order by popularity DESC limit 3')
cur.fetchall()
```

```python
# What is the most awarded average rated movie since jan 1st, 2000
cur.execute("select original_title from movies where release_date > '2000-01-01' order by vote_average DESC limit 1")
cur.fetchall()
```

```python
# Which movies were directed by Brenda chapman ? 
cur.execute("select m.original_title,d.name from movies as m join directors as d on m.director_id=d.id where name='Brenda Chapman'")
cur.fetchall()
```

```python
# Name the director who has made more number of movies
cur.execute("select d.name from directors as d inner join movies as m on d.id=m.director_id group by d.name order by count(d.name) DESC limit 1")
cur.fetchall()
```

```python
# Name the director who is most bankable
cur.execute("select d.name,sum(m.budget) from movies as m join directors as d on d.id=m.director_id group by d.name order by sum(m.budget) DESC limit 1")
cur.fetchall()
```

**<h1>Director Analysis</h1>**

```python
# Name all the directors with the number of movies total revenue generated by those total movies of that director, The director that has generated highest revenue which come at top
cur.execute("select d.name,count(m.id),sum(m.revenue) from movies as m join directors as d on m.director_id=d.id group by d.name order by sum(m.revenue) DESC ")
cur.fetchall()
```

```python
# Name all the directors with the number of movies total revenue generated by those total movies of that director, The director that has made highest movies which come at top
cur.execute("select d.name,count(m.id),sum(m.revenue) from movies as m join directors as d on m.director_id=d.id group by d.name order by sum(m.id) DESC")
top=pd.DataFrame(cur.fetchall(),columns=['Name','Total_Movies','Total_revenue'])
top
```
